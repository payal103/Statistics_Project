```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder


# Load the dataset
df = pd.read_csv('netflix_movies_series_final.csv')
df.head()
```

```python
IMDB_Rating = df["IMDB Rating"].mean()
print("IMDB Rating:",IMDB_Rating )
```

```python
mean_budget = df["Budget (Million $)"].mean()
print("Mean Budget (Million $):", mean_budget)
```

```python
mean_revenue  = df["Revenue (Million $)"].mean()
print("Revenue (Million $):", mean_revenue)
```

```python
mean_watch_Time = df["Avg User Watch Time (Minutes)"].mean()
print("Avg User Watch Time (Minutes):",mean_watch_Time)
```

```python


# Creating separate DataFrames for each metric
df_imdb = pd.DataFrame({"Category": ["IMDb Rating"], "Mean Value": [IMDB_Rating]})
df_budget = pd.DataFrame({"Category": ["Budget"], "Mean Value": [mean_budget]})
df_revenue = pd.DataFrame({"Category": ["Revenue"], "Mean Value": [mean_revenue]})
df_watch_time = pd.DataFrame({"Category": ["Watch Time"], "Mean Value": [mean_watch_Time]})

# List of DataFrames and their respective colors
data_frames = [df_imdb, df_budget, df_revenue, df_watch_time]
titles = ["Mean IMDb Rating", "Mean Budget", "Mean Revenue", "Mean Watch Time"]
colors = ["Blues", "Greens", "Reds", "Purples"]  # Different color palettes

# Plot each bar chart separately with a different color
plt.figure(figsize=(10, 8))

for i, (df_plot, title, color) in enumerate(zip(data_frames, titles, colors), 1):
    plt.subplot(2, 2, i)  # Creating a 2x2 grid of plots
    sns.barplot(data=df_plot, x="Category", y="Mean Value", palette=color)
    plt.title(title)
    plt.ylabel("Mean Value")
    plt.xlabel("")  # No need to label x-axis individually

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

## **Statistical Summary & Key Insights**

### **1. Central Tendency (Mean, Median, Mode)**
- **Mean (Average):** Represents the expected value for IMDb Rating, Budget, Revenue, and Watch Time.
- **Median (Middle Value):** Shows the central value, less affected by extreme highs or lows.
- **Mode (Most Frequent Value):** Indicates the most common occurrence in each category.

 **Interpretation:**  
- If **Mean  Median**, the distribution is **symmetrical**.  
- If **Mean > Median**, it's **positively skewed** (right-tailed).  
- If **Mean < Median**, it's **negatively skewed** (left-tailed).  

---

### **2. Dispersion (Variance, Range, Standard Deviation, Mean Deviation)**
- **Variance:** Measures how spread out the data points are.
- **Range:** The difference between maximum and minimum values.
- **Standard Deviation (Std Dev):** The average deviation from the mean.
- **Mean Deviation:** The average absolute deviation from the mean.

 **Interpretation:**  
- A **high variance** or **standard deviation** suggests **inconsistent values** (large fluctuations).  
- A **low variance** suggests **stable** and **consistent** values.

---

### **3. Shape of Distribution (Skewness & Kurtosis)**
- **Skewness:** Measures asymmetry.
  - **Positive Skew (> 0):** Right-skewed (long right tail, few extreme high values).
  - **Negative Skew (< 0):** Left-skewed (long left tail, few extreme low values).
  - **Zero Skew ( 0):** Symmetrical distribution.

- **Kurtosis:** Measures peak and tail behavior.
  - **High Kurtosis (> 3):** Heavy tails, more extreme values (outliers).
  - **Low Kurtosis (< 3):** Light tails, few extreme values.
  - **Normal Kurtosis ( 3):** Normal distribution.

 **Interpretation:**  
- If Revenue has **high positive skew**, a few movies generate exceptionally **high revenue** while most earn lower amounts.  
- If IMDb Rating is **slightly negative skewed**, more movies tend to have higher ratings.
